### Adding a New Column

```python
df_new = df.copy()

# Converting Total_Population from object to numbers 
if df_new['Total_Population'].dtype == 'object':
    df_new['Total_Population'] = df_new['Total_Population'].str.replace(',', '').astype(float)

# Removing the % from Population_Growth_Rate and turning it into a float
if df_new['Population_Growth_Rate'].dtype == 'object':
    df_new['Population_Growth_Rate'] = df_new['Population_Growth_Rate'].str.replace('%', '').astype(float) / 100
else:
    # This section makes sure that it is a percentage by being less than 1
    if df_new['Population_Growth_Rate'].any() > 1:
        df_new['Population_Growth_Rate'] = df_new['Population_Growth_Rate'] / 100

df_new['Predicted_Population_Next_Year'] = df_new['Total_Population'] * (1 + df_new['Population_Growth_Rate'])
print(df_new)
df_new.head()
```

To create a new column. I turned both the total population and population growth rates into floats and removed symbols such as "," and "%" that made them strings. The new column is the predicted population of next year by adding 1 to the population growth rate and then multiplying that by the total population. An example would be total population= 6, growth rate=0.03, (1+0.03)=1.03 * 6= 6.18. The predicted population for the next year based on those numbers would be 6.18. Something I can explore further is how to take it out of the equation and just display the real number.
