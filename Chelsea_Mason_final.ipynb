### Cleaning Data

```python
#Checking which columns have missing data
missing_data = df.isnull().sum()
print(missing_data)
```

Every column is filled there is no missing data. That is a very pleasant suprise. Since there's no missing data there is no need to delete any columns.  

```python
#Checking for duplicates
duplicated_data = df.duplicated().sum()
print(f'Number of duplicates:{duplicated_data}')
```

There's no duplicates. There is no need for further editing in this department. The dataset is pretty much ready for analysis.

```python
#Creating a copy of the data
df_new=df.copy()
#Converting year to Categorical 
df_new['Year']=df_new['Year'].astype('category')
#Creating a new data set with Year as a Category
df_new.to_csv("new_housing_market.csv")
```

The year was turned into a category so graphs can use the year as listed instead of viewing it as a numerical value. Now graphs can compare years. A new dataset was created, it is pretty much the same as the original because the original had no duplicates or missing values. The only real difference is that the Year in the new csv file will be counted as a category. To reiterate there was no further editing of the dataset besides making the "Year" column a category because there were no missing values and no duplicates.

```python
#Checking if Year is officially a category
df_new.info()
```

It worked the Year is now officially listed as a category.

```python
#double checking for any outliers by creating a boxplot
df_new.boxplot(figsize=(12,6))
plt.title("Boxplot of Columns")
plt.xticks(rotation=45)
plt.show()
```

### Required Visualizations

```python
#Establishing correlation
corr=df_new.corr(numeric_only=True)
#Creating a Heat Map
plt.figure(figsize=(16, 14)) 
sns.heatmap(corr, annot=True, cmap='PiYG', linewidths=0.5)
plt.title('Correlation Heatmap of Housing Dataset')
plt.savefig("Correlation_Heatmap_Housing.png")
plt.show()
```

The creation of this is to view correlations. There aren't really any strong correlations. The strongest for a positive correlation are GDP Growth % and Mortgage Rate % at 0.12, and Population Growth % and Inflation Rate % at 0.12, Rent Index and Price Index at 0.098. The strongest for negative correlation are House Price Index and Affordability Ratio at -0.13, Urbanization Rate and Affordability Ratio -0.18, and Construction Index and House Price Index -0.11.

```python
#Creating a Bar Plot
plt.figure(figsize=(12,6))
sns.barplot(
    df_new.groupby("Country")["House Price Index"].mean().reset_index(),
    x="House Price Index",
    y="Country",
    palette="Blues"
)
#Labeling
plt.title("Average House Price Index by Country")
plt.xlabel("Average House Price Index")
plt.ylabel("Country")
plt.savefig('House_Price_Index_by_Country.png')
plt.show()
```

House Price Index tracks the changes in house prices over time. It shows changes in home values. The barplot shows the average House Price Index per country. Looking at the plot most countries Indexes are over 120, the two countries that fall below it are Canada and South Africa. A low House Price Index indicates homes are more inexpensive compared to the median income of that country. This implies that houses are more affordable in Canada and South Africa. Some of the highest House Price Indexes are above 140. Those countries are Brazil, Netherlands, and Switzerland. A high House Price Index indicates homes are more expensive compared to the median income of that country. Housing in these three countries are probably pretty pricey. Most of the countries fall in between 120 to 140 because of this it makes most of the countries look like they are closer to being on the expensive end when it comes to home prices.

```python
#Creating A Mortgage rate histogram
Mortgage = df_new[['Mortgage Rate (%)']] 
sns.histplot(Mortgage['Mortgage Rate (%)'], bins=12)
plt.title('Mortgage Rate Percentage Distribution')
plt.savefig("Histogram_Mortgage_Rate_Percentage.png")
plt.show()
```

The Mortgage Rate percentage are lower numbers with decimals points. This can be seen in the graph due to multiple bars being in between whole numbers. The highest count of the Mortgage Rate % is in between 4 to 5. It raises the question of why so many countries have mortgages between 4 and 5. The lowest count is in between 2 and 3, it is the only one under the count of 10. Most of the different levels of Rates are above 10. There are four in between 10 to 15 and seven above 15.   Based on the highest counts being between 4 to 5 it looks like a good portion of countries pay that Mortgage Rate Percentage.

```python
#Creating a line plot
plt.figure(figsize=(10,6))
sns.lineplot(
    df_new.groupby("Year")["Affordability Ratio"].mean().reset_index(),
    x="Year",
    y="Affordability Ratio",
    marker="o"
)
plt.title("Average Affordability Ratio Over Time (20152024)")
plt.xlabel("Year")
plt.ylabel("Affordability Ratio")
plt.savefig("Lineplot_Affordability_Ratio.png")
plt.show()
```

The Affordability Ratio is a metric used to measure a household's ability to afford things relating to goods, services, and housing. A high Affordability Ratio means that housing is less affordable. Whereas a low affordability ratio means it is easier to afford things. It is notable that the lowest the Affordability Ratio was, was in 2024 being at 6.5 for Affordability Ratio. It makes one wonder what was lowering prices globally during 2024. The year 2019 was the most expensive just slightly below 8. But after 2019 there is a steady decline of the Affordability Ratio until 2024. The year 2024 was a drop compared to 2023. The global event that comes to mind that may have impacted affordabilty Ratio around 2019 and 2020 was the Covid-19 Pandemic. However, this dataset does not have columns for world events, so it is harder to connect the two events, but speculations can be made.  Another aspect to consider is the incline from 2018 to 2019, it goes from just under 6.8 to just under 8. To summarize the line plot 2015 to 2017 declines from under 7.6 to around 6.7. Then there is an incline 2017 to 2019, from around 6.7 to slightly under 8. Then there's a drop from 2019 to 2024, from under 8 to 6.5. This entire line plot implies that the pendulum swings both ways. Houses may be affordable for a couple of years then get expensive for some time then go back to being cheaper.

```python
#Grouping House Price Index and Country
#Getting the average House Price Index for a Country
avg_hpi = df_new.groupby("Country")["House Price Index"].mean().reset_index()
#Creating a Geo Map
fig = px.choropleth(
    avg_hpi,
    locations="Country",
    locationmode="country names",
    color="House Price Index",
    color_continuous_scale="Blues",
    title="Geographic Distribution of Average House Price Index"
)
plt.savefig("GeoMap_HPI.png")
fig.show()
```

Looking at the Geo Map the Highest House Price Index belongs to Brazil, the Netherlands, and Switzerland. A high House Price Index indicates homes are more expensive compared to the median income of that country. Brazil, the Netherlands, and Switzerland are very expensive in regards to housing. A low House Price Index indicates homes are more inexpensive compared to the median income of that country. Canada and South Africa are the cheapest by this measure. One of the issues with using a Geo Map is that it is harder to see the color of some of the smaller countries. Looking at this map in order to see Europe clearly it needed to be zoomed in. This dataset only has 20 countries so the countries that are colored on the map are very few. But this allows one to better visually compare Countries. The cluster of darker countries seem to be in Europe indicating higher prices for buying homes. 

### Additional Visualizations

```python
# Calculate average House Price Index for each country
avg_by_country = df_new.groupby('Country')['House Price Index'].mean().reset_index()

# Sort and get top 5 countries
top_n = 5
top_countries = avg_by_country.sort_values('House Price Index', ascending=False).head(top_n)
top_country_list = top_countries['Country'].tolist()

# Filtering the dataframe to include only these top countries
filtered_df = df_new[df_new['Country'].isin(top_country_list)]

# Creating a bar plot showing affordability ratio by country and year
plt.figure(figsize=(16, 10))
sns.barplot(x='Country', y='House Price Index', hue='Year', data=filtered_df, palette='viridis')
plt.title('Top Countries by House Price Index', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('House Price Index', fontsize=12)
plt.savefig("Top_Countries_by_HPI.png")
plt.show()

```

The Top Countries by House Price Index are the most expensive. This graph was made to examine both the years and countries when it comes to House Price Index. For most of these countries it seems that 2024 was the more expensive year except for Brazil and Switzerland. It seems that in more recent years Switzerland has lowered their prices. France and Italy seems to have a more consistent incline of raising costs compared to the other countries. 2015 seems like a notable year for Italy, Brazil, and the Netherlands because they all seem to be higher for the countries because afterwards there is a drop. 

```python
# Calculate average House Price Index for each country
avg_by_country2 = df_new.groupby('Country')['House Price Index'].mean().reset_index()

# Sort and get bottom 5 countries
bottom_n = 5
low_countries = avg_by_country2.sort_values('House Price Index', ascending=False).tail(top_n)
low_country_list = low_countries['Country'].tolist()

# Filter the original dataframe to include only these bottom countries
df_filtered= df_new[df_new['Country'].isin(low_country_list)]

# Create a grouped bar plot showing House Price Index by country and year
plt.figure(figsize=(16, 10))
sns.barplot(x='Country', y='House Price Index', hue='Year', data=df_filtered, palette='plasma')
plt.title('Lowest Countries by House Price Index', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('House Price Index', fontsize=12)
plt.savefig("Bottom_Countries_by_HPI.png")
plt.show()
```

For Bottom House Price Index the most part there is less frequent change with these countries as compared to the top House Price Index. It seems however that prices increased for most of these countries in 2024 except in Canada. That Japan and Sweden seemed to have a lower House Price Index around the 2020. This could potentially be explained by the Covid 19 pandemic. The lowest any of these countries reach is slightly above 75 and it is Japan.

```python
#Grouping the year with GDP% and Mortgage Rate %
#It takes the average for each year
avg_year = df.groupby("Year")[["GDP Growth (%)", "Mortgage Rate (%)"]].mean().reset_index()
#Determining the size of figure
plt.figure(figsize=(10,6))
#Creating the GDP Line
plt.plot(avg_year["Year"], avg_year["GDP Growth (%)"], 
         marker="o", label="GDP Growth (%)", color="blue")
#Creating the Mortgage Rate Line
plt.plot(avg_year["Year"], avg_year["Mortgage Rate (%)"], 
         marker="o", label="Mortgage Rate (%)", color="red")
#Labeling 
plt.title("GDP Growth vs Mortgage Rate Over Time (20152024)")
plt.xlabel("Year")
plt.ylabel("Value (%)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('GDP_Growth_vs_Mortgage_Rate.png')
plt.show()
```

The GDP Growth % indicates a change in the value of goods and services produced. A high GDP Growth % shows that a countrys economy is doing well or experiencing a boost. A low GDP Growth % shows an economy is in decline. A Mortgage Rate is the percentage of a loan someone pay to the loan lender in order to own their home.  The reason why this chart was created is that there is a correlation between GDP Growth % and Mortgage Rate % at 0.12. Looking at the line chart it is clear that the lines mirror each other to some extent, or is an inverse. When the GDP increases the Mortgage Rate decreases. When the GDP decreases it seems the Mortgage Rate increases. Some areas of the lines correlation are harder to see than others like in 2017 to 2018. That time frame seems to be when they both decrease. The lines do not intersect at any point. The Mortgage Rate is higher overall than the GDP Growth %. It seems that something good was happening in 2024 because the GDP Growth % goes up and the Mortgage Rate % goes down. That means in 2024, more countries experienced a growing economy, and their Mortgage Rates decreased. That seems like a great year.
